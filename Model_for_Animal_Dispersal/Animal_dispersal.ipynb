## Fisher-Kolmogorov equation with density dependant diffusion.

Welcome back everybody! In this notebook we will study a non-linear differential equation introduced, separately, in 1937 by R. A. Fisher and A. N. Kolmogorov. We are not only introduce this equation, but also we will analyze how this equation change if we include density dependant diffusion. 

The Fisher-Kolmogorov equation is the natural extension of the logistic growth population model. Fisher suggested it as a deterministic version of a stochastic model for the spatial spread of a favoured gene in a population, and Kolmogorov study this equation, let's say in a mathy way. It is the classic simplest case of a nonlinear reaction diffusion equation, in 1D:

\begin{equation}
        \frac{\partial u}{\partial t}= au\left(1-\frac{u}{k}\right) + D_o\frac{\partial^2 u}{\partial x^2}
\end{equation}

where $k$ is the carrying capacity, it determines the size of the stable state population while $a$ is a measure of the rate at which $k$ is reached; that is, it is a measure of the dynamic. $D_o$ is a constant known as the diffusion coefficient.

First, we are going to introduce the Fisher-Kolmogorov equation with constant diffusion. Do you remember the diffusion equation? If not, go back to [lesson02_03](http://nbviewer.ipython.org/github/numerical-mooc/numerical-mooc/blob/master/lessons/02_spacetime/02_03_1DDiffusion.ipynb) of ["Practical Numerical Methods with Python"](http://openedx.seas.gwu.edu/courses/GW/MAE6286/2014_fall/about), and make a quick review before start. Then, we are going to introduce a density dependant diffusion problem, use to model animal dispersal; and finally, guess what?...Yes, we are going to put all together and conclude in a Fisher-Kolmogorov equation with density dependant diffusion.


### Discretizing F-K equation.

If you look at the equation (1), except for the first term in the RHS (logistic term), you should be able to recognize the other terms and how to discretize them. Yes! without the logistic term, is just the diffusion equation (Did you hear the voice saying forward-time and central-space?). If you pay attention, you will see the logistic term just involves $u$, so it should be easy for you to discretize the whole equation. Try it in a paper and see if you get this:  

\begin{equation}
    u_{i}^{n+1}=u_{i}^{n} + \Delta t\; a\;u_{i}^{n}(1-\frac{u_{i}^{n}}{k})    +\frac{D_o\Delta t}{\Delta x^2}(u_{i+1}^{n}-2u_{i}^{n}+u_{i-1}^{n})
\end{equation}

Remember, the central difference approximation of the 2nd-order derivative is 2nd-order accurate.


### What about stability?

Did you try to make a Von-Neumann analysis, wasn't it tempting? What happend?...Yes, due to we have a non-linear term we cannot make this analysis. However, we can try to use the stability condition we know for the diffussion equation and see what happend. Cross your fingers, unless you wanna blow things up. 

Making a quick review, for a set of discretization parameters $\Delta x$ and $\Delta t$, the stability condition for difussion is:

\begin{equation}
    D_o \frac{\Delta t}{\Delta x^2} \leq \frac{1}{2}
\end{equation}


### Let's code, I mean solve!

As always we need celles to initialize the problem by loading the libraries we need, also we have to define some parameters and of course the initial condition.
Don't forget we will use the CFL condition for diffusion and calculate a value of $\Delta t$ that will be in the stable range. We choose some parameters, but definitely we encourage you to try different ones.

```python
import numpy as np
from JSAnimation.IPython_display import display_animation
from matplotlib import animation
import matplotlib.pyplot as plt    
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 14
```

```python
#Setting initial values and parameters

nx = 161                     #number of points of space grid
L=5.                         #Half Length of the space. Space (-L,L)
dx = 2.*L/(nx-1)             #size spatial step, the 2. is cause the space is twice L

nt = 200                     #number of time steps
sigma = .2                   #CFL condition less than 0.5
Do=0.05                      #Diffusion coeficient
dt = sigma*dx**2/Do          #defensively size of time step. 

k=60.                        #Carrying capacity
a=50.                        #Measure of the rate at which k is reached
uo=10.                       #Initial density.
```

```python
#Setting initial condition

x = np.linspace(-L,L,nx)
u = np.zeros(len(x))      

#We want a hat function centered in zero but not too wide, so inspecting x,we want:

b=np.where(x==-0.125)
c=np.where(x==0.125)

#print b,c  #Uncomment this line to know where are b and c in x array.

u[78:82+1]=uo
un = np.zeros(nx)

#Plotting the initial condition

plt.xlim(-5.,5.)
plt.xticks(np.arange(-5,5.1,1))
plt.ylim(0,10.5)
plt.plot(x,u)
plt.xlabel('Distance', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(True)
plt.title('Initial Condition', fontsize=16);
```

```python
#Let's define the function to solve the Fisher-Kolmogorov equation
#It's a non-linear reaction-diffusion equation so the name will be nl_reac_diff

def nl_reac_diff(u, Do, a, k, Nt, dt, Nx, dx):
    
    """Solves the Fisher-Kolmogorov equation with forward difference in time and
    central in space.
    
    Paremeters:
    ----------
    u:  array
        initial condition

    Do, a, k:  float
               parameters of the F-K equation

    Nt: integer
        number of time steps
        
    dt: float 
        size of time step 

     Nx: integer
         number of space steps

    dx: float
        size of spatial mesh

    Returns:
    -------
    un: array of float
        Density of the Population
        
    """
    un = np.zeros((Nt,Nx))
    un[:,:] = u.copy() 
    
    for t in range(1,Nt):
        
        un[t,1:-1] = u[1:-1]+dt*a*u[1:-1]*(1-u[1:-1]/k) + Do*dt/dx**2*((u[2:] -2*u[1:-1] +u[0:-2])) 

        u = un[t,:].copy()
        
    return un
```

```python
nl_reac_diff=nl_reac_diff(u, Do, a, k, nt, dt, nx, dx)
```

```python
#Doing the animations!!!

fig = plt.figure();
ax = plt.axes(xlim=(-5,5),ylim=(0,65.5),xlabel=('Distance'),ylabel=(' Population density'));
plt.xticks(np.arange(-5,5.1,1))
line, = ax.plot([],[],color='#003366', lw=2);

def animate(frame):
    x = np.linspace(-L,L,nx)
    y = nl_reac_diff[frame,:]
    line.set_data(x,y)
    return line,

anim = animation.FuncAnimation(fig, animate, frames=nt, interval=50)
display_animation(anim, default_mode='once')
```
